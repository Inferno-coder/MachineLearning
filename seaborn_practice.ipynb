#Seaborn Practice

importing the necessory libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')
df.head()
```

```python
bar=sns.barplot(x='gear',y='mpg',data=df,palette="rocket")
```

```python
  sns.countplot(x='cyl',data=df)
```

```python
#horizontal count plot
sns.countplot(y='gear' ,data=df,color='b')
```

```python
#count of the cylinders according to the gear
sns.countplot(x='cyl',hue='gear',data=df)
```

```python
 sns.distplot(df.mpg,bins=5,color='g')
```

```python
sns.heatmap(df.corr(),cbar=True)
```

iris Dataset

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
 sns.countplot(x='species',data=iris)
```

```python
sns.heatmap(iris.corr())
```

```python
sns.scatterplot(x='petal_length',y='petal_width',data=iris,hue='species')
```

```python
 sns.pairplot(iris,hue='species',palette='Set1')
```

```python
 sns.lmplot(x='petal_length',y='petal_width',data=iris)
```
